```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title Load the dataset with the correct encoding
education_data = pd.read_csv('/content/Global_Education.csv', encoding='latin1')
```

```python
# @title Explore the dataset
print(education_data.head())  # Display the first few rows of the dataset
print(education_data.info())  # Get information about the dataset, check for missing values
```

```python
# @title 1. Horizontal Bar Chart: Out-of-School Rates by Gender in Primary Age
# Select a subset of countries for better clarity (you can customize this)
selected_countries = education_data['Countries and areas'].head(25)

# Filter the dataset
subset_data = education_data[education_data['Countries and areas'].isin(selected_countries)]

# Horizontal bar chart for out-of-school rates by gender
plt.figure(figsize=(12, 8))
sns.barplot(x='OOSR_Primary_Age_Male', y='Countries and areas', data=subset_data, color='blue', label='Male')
sns.barplot(x='OOSR_Primary_Age_Female', y='Countries and areas', data=subset_data, color='red', label='Female')
plt.title('Out-of-School Rates by Gender in Primary Age')
plt.xlabel('Out-of-School Rate')
plt.ylabel('Countries and Areas')
plt.legend()
plt.show()
```

```python
# @title Grouped bar chart for literacy rates among male and female youths
plt.figure(figsize=(12, 8))
bar_width = 0.35

# Select a subset of countries for better clarity (you can customize this)
selected_countries = education_data['Countries and areas'].head(5)

# Filter the dataset
subset_data = education_data[education_data['Countries and areas'].isin(selected_countries)]


# Calculate positions for bars
bar_positions_male = range(len(selected_countries))
bar_positions_female = [pos + bar_width for pos in bar_positions_male]

# Plot bars
plt.bar(bar_positions_male, subset_data['Youth_15_24_Literacy_Rate_Male'], width=bar_width, label='Male')
plt.bar(bar_positions_female, subset_data['Youth_15_24_Literacy_Rate_Female'], width=bar_width, label='Female')

# Customize the plot
plt.xlabel('Countries and Areas')
plt.ylabel('Literacy Rate')
plt.title('Literacy Rates Among Youths (Aged 15-24)')
plt.xticks([pos + bar_width / 2 for pos in bar_positions_male], selected_countries)
plt.legend()
plt.show()
```

```python
# @title Histogram for distribution of proficiency in reading at the end of primary education
plt.figure(figsize=(10, 6))
plt.hist(education_data['Primary_End_Proficiency_Reading'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Proficiency in Reading at the End of Primary Education')
plt.xlabel('Proficiency in Reading')
plt.ylabel('Frequency')
plt.show()
```

```python
# @title Box plot for proficiency in math at the end of lower secondary education
plt.figure(figsize=(10, 6))
sns.boxplot(x=education_data['Lower_Secondary_End_Proficiency_Math'])
plt.title('Box Plot: Proficiency in Math at the End of Lower Secondary Education')
plt.xlabel('Proficiency in Math')
plt.show()
```

```python
# @title Scatter plot for Birth Rate vs. Gross Tertiary Education Enrollment
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Birth_Rate', y='Gross_Tertiary_Education_Enrollment', data=education_data)
plt.title('Scatter Plot: Birth Rate vs. Gross Tertiary Education Enrollment')
plt.xlabel('Birth Rate')
plt.ylabel('Gross Tertiary Education Enrollment')
plt.show()
```

```python
# @title Stacked bar chart for enrollment in primary and tertiary education
plt.figure(figsize=(90, 50))
plt.bar(education_data['Countries and areas'], education_data['Gross_Primary_Education_Enrollment'], label='Primary')
plt.bar(education_data['Countries and areas'], education_data['Gross_Tertiary_Education_Enrollment'], bottom=education_data['Gross_Primary_Education_Enrollment'], label='Tertiary')
plt.xticks(rotation=45, ha='right')
plt.title('Enrollment in Primary and Tertiary Education')
plt.xlabel('Countries and Areas')
plt.ylabel('Enrollment')
plt.legend()
plt.show()
```

```python
# @title Pair plot to visualize relationships between key variables
sns.pairplot(education_data[['Birth_Rate', 'Unemployment_Rate', 'Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female']])
plt.suptitle('Pair Plot: Relationships Between Key Variables', y=1.02)
plt.show()
```

```python
# @title Heatmap for the correlation matrix
correlation_matrix = education_data.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
